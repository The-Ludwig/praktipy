```python
from praktipy import *
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 20)
y = np.sin(x)
yp = None
xi = np.linspace(x[0], x[-1], 100)
yi = np.interp(xi, x, y, yp)

fig, ax = plt.subplots()
plt.ticklabel_format(useLocale=True)
ax.plot(x, y, 'o', xi, yi, '.')
plt.show()
```
